#### 1. Check whether a given number is divisible by 3 and 5 both.

```python
num = int(input("enter a number:"))
if num % 3 == 0 and num % 5 == 0:
    print("number is divisible by both 3 and 5")
else:
        print("number is not divisible by both 3 and 5")
```

explaintion: the program checks the remainder when the number is divided by 3 and 5 if both the remainders is 0 then the number is divisible by both

#### 5. Check whether the quadratic equation has real roots or imaginary roots. Display the roots.

```python
import math

a = float(input("enter value of a: "))
b = float(input("enter value of b: "))
c = float(input("enter value of c: "))

d = b**2 - 4*a*c

if d > 0:
    root1 = (-b + math.sqrt(d)) / (2*a)
    root2 = (-b + math.sqrt(d)) / (2*a)

    print("roots are real and diffrent")
    print("root 1:",root1)
    print("root 2:",root2)

elif d == 0:
    root = -b / (2*a)
    print("roots are real and equal")
    print("root:",root)

else:
    print("roots are imaginary (complex)")
```

8. Print the grade sheet of a student for the given range of cgpa. Scan marks of
five subjects and calculate the percentage.
CGPA=percentage/10
CGPA range:
0 to 3.4 -> F
3.5 to 5.0->C+
5.1 to 6->B
6.1 to 7-> B+
7.1 to 8-> A
8.1 to 9->A+
9.1 to 10-> O (Outstanding)

```python
name = input("Enter Name: ")
roll = input("Enter Roll Number: ")
sapid = input("Enter SAPID: ")
sem = input("Enter Semester: ")
course = input("Enter Course: ")

print("\nEnter marks of 5 subjects:")
pds = int(input("PDS: "))
python = int(input("Python: "))
chem = int(input("Chemistry: "))
eng = int(input("English: "))
phy = int(input("Physics: "))

total = pds + python + chem + eng + phy
percentage = total / 5
cgpa = percentage / 10

if 0 <= cgpa <= 3.4:
    grade = "F"
elif cgpa <= 5.0:
    grade = "C+"
elif cgpa <= 6.0:
    grade = "B"
elif cgpa <= 7.0:
    grade = "B+"
elif cgpa <= 8.0:
    grade = "A"
elif cgpa <= 9.0:
    grade = "A+"
else:
    grade = "O (Outstanding)"

print("\n----- Grade Sheet -----")
print("Name:", name)
print("Roll Number:", roll, "SAPID:", sapid)
print("Sem:", sem, "Course:", course)
print("\nSubject Name: Marks")
print("PDS:", pds)
print("Python:", python)
print("Chemistry:", chem)
print("English:", eng)
print("Physics:", phy)
print("Percentage:", percentage, "%")
print("CGPA:", round(cgpa, 2))
print("Grade:", grade)
```

Explanation:l this program calculates percentage from five subject marks and computes cgpa using cgpa = percentage / 10. based on the cgpa range, it assigns the appropriate grade and prints the complete grade sheet.
