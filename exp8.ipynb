1. Add few names, one name in each row, in “name.txt file”.
a. Count no of names
b. Count all names starting with vowel
c. Find longest name

```python
file = open("viraj.txt","r")
names = file.read().split()

total = len(names)
print("total names:",total)

vowel_count = 0
for name in names:
    if name[0].lower() in "aeiou":
        vowel_count += 1

print("names starting with vowels:",vowel_count)

longest = max(names,key = len)
print("longest name:",longest)

file.close()
```

2. Store integers in a file.
a. Find the max number
b. c. Find average of all numbers
Count number of numbers greater than 100

```python
with open("numbers.txt","r") as file:
    nums = list(map(int, file.read().split()))

maximum = max(nums)
average = sum(nums) / len(nums)

count = 0
for n in nums:
    if n > 100:
        count += 1

print("maximum number:",maximum)
print("average:",average)
print("numbers greater than 100:",count)
```

Explanation: the program reads data from numbers.txt and extract only numberic values from it. it calculates the maximum number and average using built-in function. a loop is used to count how many numbers are greater than 100.


3. Assume a file city.txt with details of 5 cities in given format (cityname
population(in lakhs) area(in sq KM) ):
Example:
Dehradun 5.78 308.20
Delhi 190 1484
……………
Open file city.txt and read to:
a. Display details of all cities
b. Display city names with population more than 10Lakhs
c. Display sum of areas of all cities

```python
with open("createcity.txt","r") as file:
    cities = file.readlines()

total_area = 0

print("details of all cities:")
for line in cities:
    name,population,area = line.split()
    print(name,population,area)

print("\n cities with population more than 10 lakhs:")
for line in cities:
    name,population,area = line.split()
    if float(population) > 10:
        print(name)

for line in cities:
    name,population,area = line.split()
    total_area += float(area)

print("\n total area of all cities:",total_area)
```

Explanation: the program reads city data from createcity.txt and splits each line into city name,population,and area. it displays all city details and filters cities having population greater than 10 lakhs. finally it calculates the total area by adding the area of all cities.

4. Input two values from user where the first line contains N, the number of test
cases. The next N lines contain the space separated values of a and b. Perform
integer division and print a/b. Handle exception in case of ZeroDivisionError or
ValueError.
Sample input
1 0
2 $
3 1
 Sample Output:
Error Code: integer division or modulo by zero
Error Code: invalid literal for int() with base 10: '$' 3

```python
N = int(input("enter number of test cases:"))

for i in range(N):
    try:
        a,b = input().split()
        result = int(a) // int(b)
        print(result)
    except ZeroDivisionError as e:
        print("error code:",e)
    except ValueError as e:
        print("error code:",e)
```

5. Create multiple suitable exceptions for a file handling program.

```python
try:
    
    filename = input("Enter file name: ")
    
    file = open(filename, "r")
    
    content = file.read()
    
    print("File Content:\n", content)
    
    file.close()

except FileNotFoundError:
    
    print("Error: File not found.")

except PermissionError:
    
    print("Error: You don't have permission to open this file.")

except IsADirectoryError:
    
    print("Error: You entered a directory instead of a file.")

except UnicodeDecodeError:
    
    print("Error: File contains unsupported text format.")

except Exception as e:
    
    print("Some unexpected error occurred:", e)
```

Explanation: this program safely opens and reads a file using exxeption handling. different expections like FileNotFoundError,PermissionError,and UnicodeDecodeError handle specific problems that may occur during file access. the general expection block catches any unexpected errors to prevent program crash.
