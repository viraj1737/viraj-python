```python
a = float(input("enter side a:"))
b = float(input("enter side b:"))
c = float(input("enter side c:"))

s = (a+b+c)/2

area = (s* (s-a) * (s-b) * (s-c)) ** 0.5
print("area of triangle:",area)
```

9 .Write a program to convert given seconds into hours, minutes and remaining
seconds.

```python
seconds = int(input("Enter seconds: "))

hours = seconds // 3600
minutes = (seconds % 3600) // 60
remaining_seconds = seconds % 60

print("Hours:", hours)
print("Minutes:", minutes)
print("Seconds:", remaining_seconds)
```

Explanation: this program converts total seconds into hours,minutes. and remaining seconds using division and modulus operators.
it uses // for whole number division and % to get the remainder.
