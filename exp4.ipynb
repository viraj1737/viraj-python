1. Find a factorial of given number.

```python
num = int(input("enter a number:"))

fact = 1

for i in range(1,num + 1):
    fact = fact * 1

print("factorial:",fact)
```

Explanation: this program calculates the factorail of a number using the loop. it multiplies number from 1 to the given number to get the factorial.

2. Find whether the given number is Armstrong number.

```python
num = int(input("enter a number:"))

temp = num
sum=0
n = len(str(num))
 
while temp > 0:
    digit = temp % 10
    sum += digit ** n
    temp 

if sum == num:
    print("armstrong number")
else:
    print("not an armstrong number")
```

Explanation: This program checks whether a number is an armstrong number by summing each digit raised to the power of total digits. if the calculated sum equals the original number,it is an armstrong number.

4. Write a program to find if given number is prime number or not.?

```python
num = int(input("enter a number:")) 

if num > 1:
    for i in range(2,num):
        if num % i == 0:
            print("not a prime number")
            break
    else:
        print("prime number ")
else:
    print("not a prime number")
    
```

Explanation: this program checks whether a number is prime by testing divisibility from 2 to n-1. if the number has no divisiors other than 1 and itself, it is a prime number.

5. Check whether given number is palindrome or not.

```python
num = int(input("enter a number:"))
temp = num 
reverse = 0

while temp > 0:
    digit = temp % 10
    reverse = reverse * 10 + digit
    temp //= 10

if reverse == num:
    print('palindrome number')
else:
    print("not a palidrome number")
```

Explanation: this program reverses the digits of a number using a loop. if the reversed number is equal to the origianl number , it is a palidrome.

6.Write a program to print sum of digits.

```python
num = int(input("enter a number:"))
sum_digits = 0 
while num > 0:
    digit = num % 10
    sum_digits += digit 
    num
print("sum of digits:",sum_digits)
```

Explanation: this program finds the sum of digits by extracting each digit using modulus operator %. it keeps adding the digits until the number becomes 0.

7. Count and print all numbers divisible by 5 or 7 between 1 to 100.

```python
count = 0

for i in range(1,100):
    if i % 5 == 0 or i % 7 == 0:
        print(i,end=" ")
        count += 1

print("toto=al numbers =",count)
```

Explanation: this program numbers from 1 to 100 and prints those divisible by 5 or 7 using the modulus operator %. it also counts those numbers and prints the total at the end.

8. Convert all lower cases to upper case in a string.

```python
string = input("enter a string:")
result = ""

for ch in string:
    if ch.islower():
        result += ch.upper()
    else:
        result += ch

print("updated string:",result)
```

Explanation: this program checks each character of the string and converts lowercase letters to uppercase using .upper(). other characters (uppercase,numbers,symbols),remain unchanged.

9. Print all prime numbers between 1 and 100.

```python
for num in range(2,101):
    is_prime = True

    for i in range(2,int(num ** 0.5) + 1):
        if num % i == 0:
            is_prime = False
            break

    if is_prime:
        print(num,end=" ")
```

Explanation: this program checks each number from 2 to 100 and tests divisibility up to its square root. if a number has no divisors other than 1 and itself,it is printed as a prime number.
