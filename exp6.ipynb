1. Scan n values in range 0-3 and print the number of times each value has
occurred.

```python
n = int(input("enter how many values:"))
count = [0,0,0,0]

for i in range(n):
    x = int(input())
    if 0 <= x <= 3:
        count[x]==1

for i in range(4):
    print(i,"occurred",count[i],"times")
```

Explanation:in this program we stores counts of numbers 0 to 3 using a list. each input increases its index count,then occurrences are displayed.

2. Create a tuple to store n numeric values and find the average of all values.

```python
n = int(input("enter number of values:"))
nums = ()

for i in range(n):
    val = int(input())
    nums= nums+(val,)

average = sum(nums)/n
print("average = ",average)
```

Explanation: the numbers are stored in a tuple by adding each value. the average is found by dividing the sum of all tuple elements by n.

3. WAP to input a list of scores for N students in a list data type. Find the score of
the runner-up and print the output.
Sample Input
N = 5
Scores= 2 3 6 6 5
Sample output
5
Note: Given list is [2, 3, 6, 6, 5]. The maximum score is 6, second maximum is
5. Hence, we print 5 as the runner-up score.

```python
n = int(input("enter number of students:"))
scores = list(map(int,input("enter scores:").split()))

scores = list(set(scores))
scores.sort()

print("runner-up score =",scores[-2])
```

Explanation: duplicate scores are removed using set(). after sorting,the second highest value becomes the runner-up score.

4. Create a dictionary of n persons where key is name and value is city.
a. Display all names
b. Display all city names
c. Display student names and cities of all students.
d. Count the number of students in each city. 

```python
n = int(input("enter number of persons:"))
d = {}

for i in range(n):
    name = input("enter name:")
    city = input("enter city:")
    d[name] = city

print("name:",d.keys())
print("cities:",d.values())

print("students and cities:")
for name,city in d.items():
    print(name,"->",city)

count={}
for city in d.values():
    count[city]= count.get(city,0)+1

print('number of students in each city:')
for city in count:
    print(city,":",count[city])
```

Explanation: a dictionary is used to store name as key and city sa value. keys and values help display names,cities,and student-city pairs. another dictionary counts how many students belong to each city 


5. Store details of n movies in a dictionary by taking input from the user. Each
movie must store details like name, year, director name, production cost,
collection made (earning) & perform the following :-
a. print all movie details
b. display name of movies released before 2015
c. print movies that made a profit.
d. print movies directed by a particular director.

```python
n = int(input("enter number of movies:"))
movies ={}

for i in range(n):
    name=input("enter movie name:")
    year=int(input("enter release year:"))
    director=input("enter director name:")
    cost=int(input("enter production cost:"))
    collection=int(input("enter collection:"))
    movies[name]={"year":year,"director":director,"cost":cost,"collection":collection}

print("all movie details:")
for m in movies:
    print(m,movies[m])

print('movies released before 2015:')
for m in movies:
    if movies[m]['year']<2015:
        print(m)

print("profitable movies:")
for m in movies:
    if movies[m]["collection"]>movies[m]["cost"]:
        print(m)

d = input("enter director name to search:")
print("movies by",d,":")
for m in movies:
    if movies[m]["director"]==d:
        print(m)
```

Explanation: movies are stored using the nested dictionay with details like year,cost and collection. conditions are used to find old movies,profitable movies and movies by a specific director. profit is when collection is greater than production cost.
