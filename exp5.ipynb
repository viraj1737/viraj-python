1. Write a program to count and display the number of capital letters in a given
string.

```python
string = input("enter a string:")
count = 0

for ch in string:
    if ch.isupper():
        count += 1

print("number of capital letters:",count)
```

Explanation: this program checks each characte of the string using .isupper() to find capital letters. it counts them and finally prints the total number of uppercase letters.

2. Count total number of vowels in a given string.

```python
string = input("enter a string:")

count = 0
vowels  = "aeiouAEIOU"

for ch in string:
    if ch in vowels:
        count += 1

print("total number of vowels:",count)
```

Explanation: this program checks each character of the string to see if it is a vowel(a,e,i,o,u). if it matches,the counter increases and finally the total number of vowels is displayed.

3. Input a sentence and print words in separate lines.


```python
sentence = input("enter a sentence:")
words = sentence.split()

for word in words:
    print(word)
```

Explanation: this program takes a sentence as input and uses .split() to  separate it into words. then it prints each word on a new line using a for loop.

4. WAP to enter a string and a substring. You have to print the number of times
that the substring occurs in the given string. String traversal will take place from
left to right, not from right to left.
Sample Input
ABCDCDC
CDC
Sample Output
2

```python
string = input("enter main string:")
substring = input("enter substring:")

count = 0 

for i in range(len(string) - len(substring) + 1):
    if string[i: i+len(substring)] ==substring:
        count += 1

print(count)
```

Explanation: this program checks each possible position in the string from left to right using slicing. if the sliced part matches the substring,the counter increases and the total is printed.

5. Given a string containing both upper and lower-case alphabets. Write a Python
program to count the number of occurrences of each alphabet (case
insensitive) and display the same.
Sample Input
ABaBCbGc
Sample Output
2A
3B
2C
1G

```python
s = input('enter string:').upper()
count = {}
for ch in s:
    if ch.isalpha():
        count[ch] = count.get(ch,0) + 1

for i in sorted(count):
    print(count[i],i,sep="")
```

Explanation: this program first converts the string to uppercase to make counting case-insensitive. then it counts each alphabet using a dictionary and prints the occurrences in sorted order. 

6. Program to count the number of unique words in a given sentence using sets.

```python
s = input("enter sentence:").lower()
words = s.split()
unique_words = set(words)
print("number of unique words =",len(unique_words))
```

Explanation: the sentence is converted to lowercase and split into words. a set automatically removes duplicates,so its length gives the number of unique words.

7. Create 2 sets s1 and s2 of n fruits each by taking input from user and find:
a. Fruits which are in both sets s1 and s2
b. c. Fruits only in s1 but not in s2
Count of all fruits from s1 and s2


```python
n = (int(input("enter number of fruits in each set:")))

print("enter fruits for set 1:")
s1 = set()
for i in range(n):
    s1.add(input())

print("enter fruits for set 2:")
s2 = set()
for i in range(n):
    s2.add(input())

print("common fruits:",s1&s2)
print("fruits only in set1:",s1-s2)
print("total fruits count:",len(s1|s2))
```

Explanation: in this program & finds common fruits, - finds fruits only in s1. | combines both sets and its length gives total unique fruits.

8. Take two sets and apply various set operations on them:
S1 = {Red, yellow, orange, blue}
S2 = {violet, blue, purple}

```python
s1 = {"red","yellow","orange","blue"}
s1 = {"violet", "blue", "purple"}

print("union:",s1 | s2)
print("intersection:",s1&s2)
print("only in s1:",s1-s2)
print("only in s2:",s2-s1)
print("symmetric difference:",s1^s2)
```

Explanation: in this program union joins both sets,intersection finds coommon elements. difference shows unique elements in each set, while symmetric difference shows non - common elements from both.
